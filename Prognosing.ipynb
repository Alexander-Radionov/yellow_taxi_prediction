Вспомогательный класс. Его использование помогает в 2-3 раза уменьшить время на формирование выборок (реализация тут - https://stackoverflow.com/questions/13264946/how-to-assign-list-of-values-to-a-key-using-ordereddict-in-python

, объяснение здесь - https://stackoverflow.com/questions/27929472/improve-row-append-performance-on-pandas-dataframes )

```python
from collections import OrderedDict

class ListOrderedDict(OrderedDict):
    def __missing__(self,k):
        self[k] = []
        return self[k]
```

Функция преобразования агрегированных по регионам данных в DataFrame, пригодный для обучения моделей.

```python
def getSample(data,passengers,distance,payment,prevHours,prevDays,postHours):
    listOrdDict = ListOrderedDict()
    q = 0
    columns = ['weekday','isHoliday','hour']
    
    for k in range(prevHours):
        columns.append('trips_-'+str(k))
        columns.append('passengers_-'+str(k))
        columns.append('distance_-'+str(k))
        columns.append('totalPay_-'+str(k))
    columns.append('tripsHoursMean')
    
    for k in range(1,prevDays+1):
        columns.append('tripsDay_-'+str(k))
        columns.append('passengersDay_-'+str(k))
        columns.append('distanceDay_-'+str(k))
        columns.append('totalPayDay_-'+str(k))
            
    for i in range(prevHours+prevDays*24+1,data.shape[0] - postHours):
        
        for j in range(102):
            
            weekday = data.index[i].weekday()
            listOrdDict[q].append(weekday)
            if weekday >= 6:
                listOrdDict[q].append(1)
            else:
                listOrdDict[q].append(0)
            listOrdDict[q].append(data.index[i].hour)
            
            tripsSum = 0.
            for k in range(prevHours):
                trips = data.iloc[i-k,j]
                listOrdDict[q].append(trips)
                listOrdDict[q].append(passengers.iloc[i-k,j])
                listOrdDict[q].append(distance.iloc[i-k,j])
                listOrdDict[q].append(payment.iloc[i-k,j])
                tripsSum += trips
            listOrdDict[q].append(float(tripsSum)/prevHours)
            
            for k in range(1,prevDays+1):
                listOrdDict[q].append(data.iloc[i-24*k,j])
                listOrdDict[q].append(passengers.iloc[i-24*k,j])
                listOrdDict[q].append(distance.iloc[i-24*k,j])
                listOrdDict[q].append(payment.iloc[i-24*k,j])
            
            q += 1
    df = pd.DataFrame.from_dict(listOrdDict,orient='index')
    df.columns = columns
    return df
```

Функция получения вектора Y (отклика) из агрегированных данных:

```python
def getTarget(data,prevHours,prevDays,postHours):
    target = []
    for i in range(prevHours+prevDays*24+1,data.shape[0] - postHours):
        for j in range(data.shape[1]):
            target.append(data.iloc[i+postHours,j])
    return target
```

Функция форирования признаков sin, cos:

```python
def getWeeklyFeatures(data,kCoeff=5,function='sin'):
    from math import sin,cos,pi
    if function == 'sin':
        for k in range(kCoeff):
            data['sin_'+str(k+1)+'_weekly']=[sin(i*2*pi*(k+1)/168) for i in range(data.shape[0])] 
    else:
        for k in range(kCoeff):
            data['cos_'+str(k+1)+'_weekly']=[cos(i*2*pi*(k+1)/168) for i in range(data.shape[0])] 
```
